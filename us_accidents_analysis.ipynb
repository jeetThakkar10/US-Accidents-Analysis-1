```python
unique_cities = df.City.unique()    #Found the unique cities
len(unique_cities)    # Found the number of unique cities
```

```python
cities_by_accidents = df["City"].value_counts().sort_values(ascending=False)    # made a variable which stores the number of accidents occured in each city
cities_by_accidents[:20]    # Top 20 cities by accidents    
```

```python
sns.set_style("darkgrid")
```

```python
sns.distplot(cities_by_accidents)
```

```python
# Grouping cities into High Accident Cities and Low accident cities
high_accident_cities = cities_by_accidents[cities_by_accidents > 1000]   
low_accident_cities = cities_by_accidents[cities_by_accidents < 1000]
print(high_accident_cities.head(20))
print(len(high_accident_cities))
```

```python
print(len(high_accident_cities)/len(unique_cities))
print(len(low_accident_cities)/len(unique_cities))

# Insights :
# Less than 3% cities have more than 1000 yearly accidents
```

```python
sns.distplot(high_accident_cities)
```

```python
sns.distplot(low_accident_cities)
```

```python
# These histplots will represent the number of cities on Y-axis and Number of accidents on X in a log-scale
sns.histplot(high_accident_cities, log_scale = True)
plt.show()
sns.histplot(low_accident_cities, log_scale = True)
plt.show()

# Insights:
# 100+ cities have seen 1000+ accaccidents. Appears that the mode is around 1000. And we get the same value at 
print(high_accident_cities.mean())
print(high_accident_cities.median())
print(high_accident_cities.mode().mean())
```

### We will now analyse start time

```python
df['Start_Time']=pd.to_datetime(df['Start_Time'])
df['Start_Time'].tail()
```

```python
date_time = pd.to_datetime(df.Start_Time, infer_datetime_format= "%Y-%m-%d %H:%M:%S")
unique, counts = np.unique(date_time.dt.year, return_counts=True)
print(unique, counts)
sns.barplot(x=unique, y=counts)    # This, including previous 2 lines can be used if we want to plot a barplot. 
# sns.color_palette("Paired", 7)
# sns.histplot(date_time.dt.year, bins=5, palette = "Paired")
```

```python
sns.histplot(date_time.dt.hour, bins=24)
time, accidents_num = np.unique(date_time.dt.hour, return_counts=True)
max = accidents_num.max()
for i in range(len(accidents_num)):
  if accidents_num[i]== max:
    break
max_accidents_time = time[i]
print(max_accidents_time)

''' 
Insight:
maximum accidents occur in the evening time, around 5pm (Probably people goung back home from work)
'''
```

```python
# We will see on which days are the accidents higher?
date_time = pd.to_datetime(df.Start_Time, infer_datetime_format= "%Y-%m-%d %H:%M:%S")
date_time.dt.dayofweek

sns.histplot(date_time.dt.dayofweek, bins=7)


# Inferences:
# The number of accidents that occur on weekends are comparatively lower than those on weekdays.
```

Is the distribution of accidents by hour the same on weekends as on weekdays

```python
for i in range(7):
  start_time = df.Start_Time[date_time.dt.dayofweek == i]
  sns.histplot(start_time.dt.hour).set(title= "Day of the week = {}".format(i))
  plt.show()

# We just tried checking if there is any variation in accident times on weekdays and weekends. But it appears almost the same. 
```

```python
sns.histplot(date_time.dt.month, bins=12)

# Inference:
# The number of accidents occuring start increasing from month #7 and make a peak in month 12
```

```python
# WE will just look into data of 2019
sns.histplot(date_time.dt.month[date_time.dt.year==2019], bins=12)

# Inference :
# The trend looks quite balanced. Because much data is missing for 2017,18 
```
